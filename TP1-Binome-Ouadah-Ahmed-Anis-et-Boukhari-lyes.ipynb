```python
# Objective: 
#This first practical session aims to introduce you to the Python programming language.
#We will cover the installation of Python, the integration of libraries, module importation, 
#and file manipulation.
import pandas as pd 
df = pd.read_csv(r"/Users/anisouadah/anaconda3/titanic-passengers.csv",delimiter=';') 
df.isnull().sum()
#The function 'isnull().sum()' It is used to count the number of null values (or NaN) in each column of a DataFrame 
```

```python
#Use the 'info' method to display information about the DataFrame 'df'
df.info
```

```python
number_of_elements = len(df["Cabin"])
print ("Number of elements: ",number_of_elements)
#Number of elements per category 
print(df["Cabin"].value_counts())
#Replace and display the values 
```

```python
# The 'fillna' method is used to replace missing (NaN) values in the 'Cabin' column with 'G6'.
# Fill missing values in the 'Cabin' column with the value 'G6'
# The use of the inplace=True argument indicates that the modification should be performed directly on the DataFrame.
df["Cabin"].fillna('G6', inplace=True)
df.tail()
df
# The function tail()  is employed to display the last five rows of the DataFrame.
```

```python
# Use the 'value_counts()' method to count and display the frequency of unique values in the 'Age' column
print(df["Age"].value_counts())
```

```python
#The 'fillna' method is used to replace missing (NaN) values in the 'Age' column with '29.699118'.
#fill missing values in the 'Age' column with the values '29.699118'
# The use of the inplace=True argument indicates that the modification should be performed directly on the DataFrame.
df["Age"].fillna('29.699118',inplace=True)
df.tail()
#we use the function tail() to display the last five rows of the modified DataFrame.
```

```python
# Use the 'value_counts()' method to count and display the frequency of unique values in the 'Embarked' column
print(df["Embarked"].value_counts())
```

```python
#The 'fillna' method is used to replace missing (NaN) values in the 'Embarked' column with 'S'.
#fill missing values in the 'Embarked' column with the values 'S'
# The use of the inplace=True argument indicates that the modification should be performed directly on the DataFrame.
df["Embarked"].fillna('S', inplace=True)
df.tail()
df
#we use the function tail() to display the last five rows of the modified DataFrame.
```

```python
# now we are gonna do the historigram for the three categories 'Age , Embarked , Cabin'
#we import plt from the library matplotlib.pyplot
import matplotlib.pyplot as plt
# set the title of the plot
plt.title("Histogram of different ages")
#set the label for x abscis  
plt.xlabel("Age")
# create a historigramm plot for 'Age'
df['Age'].plot.hist()
```

```python
# Import the Seaborn library and alias it as 'sns'
import seaborn as sns
# Create a countplot of the 'Embarked' column using data from the DataFrame 'df'
sns.countplot(x='Embarked',data=df)
# Rotate the x-axis labels 
plt.xticks(rotation=-45)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style="darkgrid")  # Set Seaborn style

# Assuming df is defined with a 'Cabin' column
sns.countplot(x='Cabin', data=df)
plt.xticks(rotation=-45)  # Rotate x-axis labels for better readability
plt.xlabel("Cabin")
plt.ylabel("Count")
plt.title("Count of Passengers in Each Cabin")

# Adjust the y-axis scale (limits) as needed
plt.ylim(0, 5)  # Set the y-axis limits (e.g., from 0 to 50)

plt.show()  # Display the plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")  # Set the Seaborn style to 'whitegrid'

# Création of two separate historigramme , one for one sexe
plt.figure(figsize=(12, 6))  # Create a figure with a specified size

# Create the first subplot for male passengers
plt.subplot(1, 2, 1)  
sns.histplot(data=df[df['Sex'] == 'male'], x='Age', kde=True, color='blue')
plt.xlabel("Âge")
plt.ylabel("Nombre d'individus")
plt.title("Hommes")
# Create the second subplot for female passengers
plt.subplot(1, 2, 2)  
sns.histplot(data=df[df['Sex'] == 'female'], x='Age', kde=True, color='red')
plt.xlabel("Âge")
plt.ylabel("Nombre d'individus")
plt.title("Femmes")

# Use 'plt.tight_layout()' to automatically adjust spacing between subplots
plt.tight_layout()  
plt.show()  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Set the Seaborn style to 'whitegrid'
sns.set(style="whitegrid")  
# Create a FacetGrid with two columns based on the 'Survived' column in the DataFrame 'df'
# Each column will represent a different value of 'Survived'
g = sns.FacetGrid(df, col='Survived', height=6)
# Map a histogram plot of the 'Age' column to the FacetGrid
g.map(plt.hist, 'Age', bins=20, color='Red')
# Set the labels for the x and y-axes
g.set_axis_labels("Âge", "Nombre d'individus")
plt.show()  
```
